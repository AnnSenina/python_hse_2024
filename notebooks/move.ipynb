```python
import requests
from bs4 import BeautifulSoup

url = "https://move.ru/kvartiry/?rooms%5B0%5D=2&rooms%5B1%5D=-1" # меняем ссылку на ваш запрос с фильтрами

response = requests.get(url) # метод get, чтобы получать ответы (responses) с сайта
response # Какое состояние ответа - выдается номер
```

```python
soup = BeautifulSoup(response.text)
```

```python
all_urls = []
for i in range(1, 3713): # верхняя граница зависит от поискового запроса, здесь количество страниц с квартирами
    all_urls.append('https://move.ru/kvartiry/?page=' + str(i) + '&rooms[]=2&rooms[]=-1&limit=30')
all_urls
```

```python
all_pages = []
for i in all_urls[:100]: # первые 100
    response = requests.get(i)
    soup = BeautifulSoup(response.text)
    for i in soup.find_all('a', {'class' : "search-item__title-link search-item__item-link"}):
        all_pages.append('https:' + i.get('href'))
```

```python
def parse(url):
  page = requests.get(url)
  soup = BeautifulSoup(page.text)

  d = {}

  d['ссылка'] = url

  for i in soup.find_all('span', {'itemprop':"name"}):
    title = i
  d['название'] = title.text.strip()

  for i in soup.find_all('li', {'class':"object-info__details-table_property"}):
    title = i.find('div', {'class':"object-info__details-table_property_name"}).text.strip()
    data = i.find('div', {'class':"object-info__details-table_property_value"}).text.strip().replace('\xa0', ' ')
    d[title] = data

  return d
```

```python
### ИЛИ (без хождения по вложенным div - код из варианта 1):

def parse(url):
  page = requests.get(url)
  soup = BeautifulSoup(page.text)
  d = {}
  d['ссылка'] = url
  for i in soup.find_all('span', {'itemprop':"name"}):
    title = i
  d['название'] = title.text.strip()
  l = []
  colnames = []
  import re
  for i in soup.find_all('li', {'class':'object-info__details-table_property'}):
      info = re.sub(r' +', r' ', i.text.strip().replace('\xa0', '').replace('\n', '')).split(': ')
      l.append(info[1])
      colnames.append(info[0])

  d2 = dict(zip(colnames, l))
  d.update(d2)
  return d
```
